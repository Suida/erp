```python
from hello import ERGraph, Entity
```

```python
person = Entity('person', ['id', 'name', 'age', 'gender',])
er = ERGraph(nodes=[person])
student = Entity('Student', ['id', 'school', 'class', 'score', 'person_id'])
er.node(student)
er.edge(student, person, 'person_id')
er
```

```python
driver = Entity('Driver', ['truck_id', 'person_id'])
```

```python
truck = Entity('Truck', ['plate_num', 'model'])
```

```python
driver.edge_to(truck, 'truck_id')
driver.edge_to(person, 'person_id')
```

```python
er.node(driver)
er.node(truck)
```

```python
cls = Entity('class', ['id', 'start'])
cls_student_through = Entity('cls_student_through', ['student_id', 'cls_id'])
cls_student_through.edge_to(cls, 'cls_id')
cls_student_through.edge_to(student, 'student_id')
er.node(cls)
er.node(cls_student_through)
```

```python
teacher = Entity('teacher', ['id', 'person_id'])
teacher.edge_to(person, 'person_id')
cls_teacher_through = Entity('cls_teacher_through', ['cls_id', 'teacher_id'])
cls_teacher_through.edge_to(cls, 'cls_id')
cls_teacher_through.edge_to(teacher, 'teacher_id')
er.node(teacher)
er.node(cls_teacher_through)
```
